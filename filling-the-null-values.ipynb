# Fill NaN

# Replaces the NaN with Zero

```python
df.fillna(0)
```

# Filling the NaN in individual Column

```python
df.fillna({'Physics':'none','Chemistry':30,'Maths':35})
```

# forward fill method

```python
df.fillna(method='ffill')
```

# forward fill refering the previous column

```python
df.fillna(method='ffill',axis=1)
```

# Filling the mean value at the NaN

```python
df['Physics'].fillna(value=df['Physics'].mean())
```

```python
df
```

# Save the dataframe changes fixed

```python
df.fillna(method='ffill',inplace=True)
```
